```python
from datasets import list_datasets, load_dataset
import pandas as pd
import numpy as np
```

```python
dataset = load_dataset("eli5")
```

```python
df = dataset["validation_eli5"].data.to_pandas()[["q_id","title","selftext","answers"]]
```

```python
def reformat_tuples(answer):
    val = []
    for i,ID in enumerate(answer["a_id"]):
        val.append([ID,answer['text'][i], answer['score'][i]])
    return val

df["answers2"] = df['answers'].apply(reformat_tuples)
df =df.explode("answers2")
df = df.reset_index(drop = True)     
```

```python
split_df = pd.DataFrame(df['answers2'].tolist(), columns=['a_id', 'answer_text', 'score'])
df = pd.concat([df, split_df], axis=1)
df = df.drop(['answers','answers2'], axis=1)
```

```python
df["title"] = df["title"].apply(str.lower)
df["selftext"] = df["selftext"].apply(str.lower)
df["answer_text"] = df["answer_text"].apply(str.lower)
```

```python
words = ["what","where","why","when","how","which"]
def str_count(string, s):
    return string.count(s)
for word in words:
    df[word] = df["title"].apply(str_count, args = (word,)) + df["answer_text"].apply(str_count, args = (word,))
```

```python
def calc_prob(team,enemy):
    team_sum = sum(team)
    enemy_sum = sum(enemy)
    team_mean = np.mean(team)
    enemy_mean = np.mean(enemy)
    
    val1 = 1/(1+10**((enemy_sum-team_sum)/400))
    val2 = 1/(1+10**((enemy_mean-team_mean)/400))
    print(val1)
    print(val2)
```
